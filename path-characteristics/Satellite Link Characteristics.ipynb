```python
import glob
import os
import sys
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from collections import defaultdict
import numpy as np
import datetime
import matplotlib

continuous = glob.glob('continuous-ping/exp*')
intervals = glob.glob('interval-ping/exp*')

figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')
```

```python
def parse_ping(f):
    points = []
    with open(f) as ping_file:
         for line in ping_file:
            if line.startswith('['):
                ts = float(line[1:11])
                rtt = float(line[-11:].strip('time= ms\n'))
                points.append((ts,rtt))
    return points
```

```python

for f in continuous:
    with open(f) as ping_file:
        points = parse_ping(f)

formatter = matplotlib.dates.DateFormatter('%d %h %H:%M')

x, y = zip(*points)
d_sorted = defaultdict(list)
d_sorted_tuples = []
data_to_plot = []

for point in points:
    if datetime.datetime.fromtimestamp(float(point[0])).day != 28:
        d_sorted[datetime.datetime.fromtimestamp(float(point[0])).hour].append(point[1])

for point in points:
    if datetime.datetime.fromtimestamp(float(point[0])).day in [27]:
        d_sorted_tuples.append(point)

for item in d_sorted:
    data_to_plot.append(np.array(d_sorted[item]).astype(np.float))

a,b = zip(*d_sorted_tuples)

plt.boxplot(data_to_plot)
plt.ylabel('RTT (ms)')
plt.xlabel('Hour of Day')
plt.xticks(range(1,25), [ "{}:00".format(x) for x in range(0,25)])
plt.show()
```

```python
figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')
points = []
for f in continuous:
    points.extend(parse_ping(f))

formatter = matplotlib.dates.DateFormatter('%d %h %H:%M')

x, y = zip(*points)
d_sorted_tuples = []

d_sorted = defaultdict(list)
data_to_plot = []

for point in points:
    if datetime.datetime.fromtimestamp(float(point[0])).day in [27]:
        d_sorted_tuples.append(point)
        d_sorted[datetime.datetime.fromtimestamp(float(point[0])).hour].append(point[1])
        
for item in d_sorted:
    data_to_plot.append(np.array(d_sorted[item]).astype(np.float))

a,b = zip(*d_sorted_tuples)

a = [datetime.datetime.fromtimestamp(ts).hour for ts in np.array(a).astype(float)]

plt.scatter(a, b, s=0.5, color='black')
plt.show()
plt.savefig("RTT-hourly-scatter.jpg")
```

```python
points = [] 
data_to_plot = []

for f in intervals:
    points.extend(parse_ping(f))

formatter = matplotlib.dates.DateFormatter('%H:%M')

for i in range(0,5):
    points2 =  points[i*165:i*165+165]
    x, y = zip(*points2)
    x = [datetime.datetime.fromtimestamp(ts) for ts in np.array(x).astype(float)]
    data_to_plot.append(np.array(y).astype(np.float))
    plt.subplot(2, 1, 1)
    plt.plot_date( x, y, ms=0.7)

plt.title('RTT variation')
ax = plt.subplot(2, 1, 1)
plt.ylabel('RTT (ms)')
plt.xlabel('Time')
ax.xaxis.set_major_formatter(formatter)

plt.subplot(2, 1, 2)
plt.boxplot(data_to_plot)
plt.xticks([1, 2, 3, 4, 5], ['1 min', '5 min', '30 min', '60 min', '120 min'])
plt.ylabel('RTT (ms)')
plt.xlabel('Interval')

plt.show()

formatter = matplotlib.dates.DateFormatter('%H:%M:%S')
r =['1 min', '5 min', '30 min', '60 min', '120 min']
for i in range(0,5):
    points2 =  points[i*165:i*165+165]
    x, y = zip(*points2)
    x = [datetime.datetime.fromtimestamp(ts) for ts in np.array(x).astype(float)]
    plt.plot_date( x, y, ms=0.8, xdate=True)
    ax= plt.subplot()
    ax.xaxis.set_major_formatter(formatter)
    plt.xlabel('Interval: ' + r[i] )
    plt.ylabel('RTT (ms)')
    plt.show()
```
